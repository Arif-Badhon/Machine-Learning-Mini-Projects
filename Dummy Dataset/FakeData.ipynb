This notebook will guide us to generate Fake Dataset for Data Science and Machine Learing projects

First we will import all we need

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Declearing some variables

```python
num_pts = 10000
noise_frac = 0.07
```

```python
X1 = np.random.uniform(1, 51, num_pts).reshape(-1,1)
X2 = np.random.uniform(1, 51, num_pts).reshape(-1,1)
X3 = np.random.uniform(2, 51, num_pts).reshape(-1,1)
X4 = np.random.uniform(1, 51, num_pts).reshape(-1,1)
X5 = np.random.uniform(1, 51, num_pts).reshape(-1,1)
```

Now, we will stack all the X

```python
X = np.hstack((X1, X2, X3, X4, X5))
```

```python
Y = 1*X1 + 2*X2 + 3*X3 + 4*X4 + 5*X5
Y = Y.reshape(-1,1)
```

Now, We add fake noise to our label

```python
Ynoise = np.amax(Y)*noise_frac
Y += np.random.uniform(-Ynoise, Ynoise, num_pts).reshape(-1, 1)
```

```python
data = np.hstack((X1, X2, X3, X4, X5, Y))
fdf = pd.DataFrame(data, columns=['X1', 'X2', 'X3', 'X4', 'X5', 'Y'])
```

```python
corr = fdf.corr() #create map
```

Now we plot our data

```python
plt.figure(figsize= (12, 10))
sns.heatmap(round(corr, 3), annot= True, vmin=-1, vmax=1, cmap="YlGnBu", linewidths=.5)
plt.grid(b=True, color='#f68c1f', alpha=0.1)
plt.show()
```
